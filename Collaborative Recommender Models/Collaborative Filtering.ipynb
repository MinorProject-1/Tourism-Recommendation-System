```python
import numpy as np
import pandas as pd
```

```python
destinations_df = pd.read_csv("C:/Users/ANANYA/OneDrive/Desktop/Minor_Project/archive (1)/Expanded_Destinations.csv",usecols=['DestinationID','Name'],dtype = {'DestinationID':'int32','Name':'str'})
visitor_df = pd.read_csv("C:/Users/ANANYA/OneDrive/Desktop/Minor_Project/archive (1)/Final_Updated_Expanded_Reviews.csv",usecols=['DestinationID','UserID','Rating'],dtype={'DestinationID':'int32','UserID':'int32','Rating':'int32'})
```

```python
destinations_df.head()
```

```python
visitor_df.head()
```

```python
df = pd.merge(visitor_df,destinations_df,on='DestinationID')
df.head()
```

```python
filter_destination = df.dropna(axis=0,subset=['Name'])
```

```python
visitor_ratingCount = (filter_destination.groupby(by=['Name'])['Rating'].count().
                     reset_index().
                     rename(columns = {'Rating':'totalRatingCount'})
                     [['Name','totalRatingCount']]
                    )
visitor_ratingCount .head()
```

```python
rating_with_totalRatingCount = filter_destination.merge(visitor_ratingCount,left_on='Name',right_on='Name',how = 'left')
rating_with_totalRatingCount.head()
```

```python
pd.set_option('display.float_format',lambda x:'%.3f' % x)
print(visitor_ratingCount ['totalRatingCount'].describe())
```

```python
popularity_threshold = 50
rating_popular_place = rating_with_totalRatingCount .query('totalRatingCount >= @popularity_threshold')
rating_popular_place.head()
```

```python
rating_popular_place.shape
```

```python
places_features_df =rating_popular_place.pivot_table(index='Name',columns='UserID',values='Rating').fillna(0)
places_features_df.head()
```

```python
from scipy.sparse import csr_matrix
places_features_df_matrix = csr_matrix(places_features_df.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine' , algorithm='brute')
model_knn.fit(places_features_df_matrix)
```

```python
places_features_df_matrix.shape
```

```python
query_index = np.random.choice(places_features_df.shape[0])
print(query_index)
distance,indices = model_knn.kneighbors(places_features_df.iloc[query_index,:].values.reshape(1,-1),n_neighbors = 3)
```

```python
places_features_df.shape
```

```python
for i in range(0,len(distance.flatten())):
    if i==0:
        print('Recommendations for {0} : \n'.format(places_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i,places_features_df.index[indices.flatten()[i]],distance.flatten()[i]))
```
