### Content based recommendation system
#### Now let's make a recommendation based on the place's plot summarises given in the overview column. So if our user gives us a movie title, our goal is to recommend movies that share similar plot summaries

```python
places_df.head(1)['Place_desc']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None,
                      strip_accents='unicode', analyzer='word', token_pattern = r'\w{1,}',
                      ngram_range=(1, 3),
                      stop_words = 'english')

# Filling NaNs with empty string
places_df['Place_desc'] = places_df['Place_desc'].fillna('')
```

```python
# Fitting the TF-IDF on the 'Place_desc' text
tfv_matrix = tfv.fit_transform(places_df['Place_desc'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
# Reverse mapping of indices and place titles
indices = pd.Series(places_df.index, index=places_df['Place_Name']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwise similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the places
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar places
    sig_scores = sig_scores[1:11]

    # Place indices
    place_indices = [i[0] for i in sig_scores]

    # Top 10 most similar Places
    return places_df['Place_Name'].iloc[place_indices]
```

```python
# Testing our content-based recommendation system with the place Boat Race
give_rec('Boat Race ')
```
